```
#######################################
######## PCA on the iris dataset #######

## load the iris dataset from sklearn
from sklearn import datasets
iris  = datasets.load_iris()
```

```
## extract the features into pandas dataframe
import pandas as pd
data = pd.DataFrame(iris.data, columns = iris.feature_names, dtype='float64')
n,p = data.shape
```

```
## center the mean
import numpy as np
data = data - np.mean(data,axis=0)

## find the covariance matrix
covmat = 1/n*data.T@data
      
## check your answer
print(covmat)
print(data.cov())  ##close enough, but there is still some difference
```

```
## PCA
pca_eigval, pca_eigvec = np.linalg.eig(covmat)

# sort the eigenvalues in descending order by absolute values
idx = np.argsort(np.absolute(pca_eigval))[::-1]
sorted_eigval = pca_eigval[idx]
U = pca_eigvec[:,idx]
#np.cumsum(sorted_eigval)/np.max(np.cumsum(sorted_eigval))

#find the pca components for X
Y = data.dot(U)
Y
```
